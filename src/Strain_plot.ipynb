```python
import numpy as np

strain = []
files = np.arange(1, 12, 1)

for i in files:
    target_line = "!strainModesCoef"
    print(f"Processing file: {i}")
    with open(f'../data/Strain/distortion_{i}.txt', "r") as file:
        lines = file.readlines()

    for i, line in enumerate(lines):
        if line.strip() == target_line.strip():
            if i + 1 < len(lines):
                next_line = lines[i + 1]
                numbers = [float(num) for num in next_line.split()]
                strain.append(numbers)

strain = np.array(strain)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(np.arange(500, 1050, 50), strain*100, marker='o', linestyle='-')
plt.title("Gamma+1 Strain Modes")
plt.xlabel("Temperature (K)")
plt.ylabel("Mode Amplitude (%)")
plt.tight_layout()
plt.legend(["Mode 1", "Mode 2", "Mode 3", "Mode 4", "Mode 5", "Mode 6"])
plt.savefig("strain_modes.png")
plt.show()
```
