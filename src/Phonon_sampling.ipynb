```python
La_weight = 0.07/(0.07+0.04+0.26)
Mn_weight = 0.04/(0.07+0.04+0.26)
O_weight = 0.26/(0.07+0.04+0.26)

weights = np.array([La_weight, Mn_weight, O_weight])

def apply_weights(ps_atom, weights):
    weighted_ps = np.zeros_like(ps_atom[0])
    for i in range(len(ps_atom)):
        weighted_ps += ps_atom[i] * weights[i]
    return weighted_ps
```

```python
import matplotlib as mpl
from matplotlib.colors import ListedColormap
import matplotlib.font_manager as fm
from matplotlib import rcParams

rcParams['font.family'] = "ARIAL"

plt.figure(figsize=(8.5/2.54, 6/2.54), dpi=300)

top = mpl.colormaps['Oranges'].resampled(130)
bottom = mpl.colormaps['Greens'].resampled(130)

newcolors = np.vstack((top(np.linspace(0.7, 0.3, 130)),
                       bottom(np.linspace(0.3, 0.7, 130))))
newcmp = ListedColormap(newcolors, name='OrangeGreen')

colormap = newcmp
colors = colormap(np.linspace(0, 1, 4))

weighted_ps_1000 = apply_weights(ps_atom_1000, weights)
weighted_ps_800 = apply_weights(ps_atom_800, weights)
weighted_ps_600 = apply_weights(ps_atom_600, weights)
weighted_ps_400 = apply_weights(ps_atom_400, weights)

plt.plot( weighted_ps_1000[:, 0]*4.135667662340165*1000*1.12, weighted_ps_1000[:, 1]/np.max(weighted_ps_1000[:, 1])+1.5, label='1000 K', color=colors[3], linewidth=0.85)
plt.plot( weighted_ps_800[:, 0]*4.135667662340165*1000*1.12, weighted_ps_800[:, 1]/np.max(weighted_ps_800[:, 1])+1, label='800 K', color=colors[2] , linewidth=0.85)
plt.plot( weighted_ps_600[:, 0]*4.135667662340165*1000*1.12, weighted_ps_600[:, 1]/np.max(weighted_ps_600[:, 1])+0.5, label='600 K', color=colors[1] , linewidth=0.85)
plt.plot( weighted_ps_400[:, 0]*4.135667662340165*1000*1.12, weighted_ps_400[:, 1]/np.max(weighted_ps_400[:, 1]), label='400 K', color=colors[0] , linewidth=0.85)
plt.plot( xs, ys-0.7, label=r'DFT (Koza $et\ al.$)', color='black', linestyle='-', linewidth=0.85)


for axis in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[axis].set_linewidth(0.85) 
plt.gca().tick_params(width=0.85, labelsize=6, length=3)
plt.yticks(np.arange(-1, 4, 1))
plt.xlim(1, 83)
plt.ylim(-1, 3.2)

plt.xlabel('Energy (meV)', fontsize=7)
plt.ylabel('G(E) (a.u)', fontsize=7)
plt.legend(frameon=False, loc='upper left', fontsize=6, ncol=3)

plt.tight_layout()
plt.savefig('scaled_weighted_power_spectrum_vs_DFT.png', dpi=300)
plt.show()
```
