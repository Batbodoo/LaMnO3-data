```python
from ase.io import read
import numpy as np
import numpy.random as random

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import scipy

random.seed(50)
```

```python
atoms_1000 = read("../data/Van_Vleck/1000K_out.xyz", ":")
```

```python
Q3_arrays_1000 = np.array([[atom.arrays["Q3_1_automated"][site.index] for site in atom if site.symbol == "Mn"] for atom in atoms_1000])
Q2_arrays_1000 = np.array([[atom.arrays["Q2_automated"][site.index] for site in atom if site.symbol == "Mn"] for atom in atoms_1000])
```

```python
matplotlib.rcParams['axes.linewidth'] = 1.5

hist_data = np.array([
    i for i in
    np.array([Q3_arrays_1000[0],Q2_arrays_1000[0]]).T 
    if (type(i[0]) != type(None) and type(i[1]) != type(None))
]).T

fig, axes = plt.subplots(figsize=(4, 3))

norm=plt.Normalize(vmin=0, vmax=15)

axes.set_title('LaMnO$_3$: $E_g(Q_2,Q_3)$ plot')
interval = 0.015
hist_range = np.arange(-0.5, 0.5+interval, interval)
axes.hist2d(hist_data[0], hist_data[1],bins=(hist_range, hist_range), norm=norm, cmap='bwr')
axes.set_xlabel("$Q_3$ (Å)")
axes.set_ylabel("$Q_2$ (Å)")
axes.set_ylim(-0.5,0.5)
axes.set_xlim(-0.5,0.5)
axes.tick_params(axis='both', labelsize=8, width=1.5, length=5)
fig.patch.set_facecolor('white')
fig.patch.set_alpha(1)
cbar = plt.colorbar(axes.hist2d(hist_data[0], hist_data[1],bins=(hist_range, hist_range), norm=norm)[3])
cbar.ax.set_ylabel('Counts')
cbar.ax.tick_params(labelsize=8, width=1.5, length=5)

H, xedges, yedges = np.histogram2d(hist_data[0], hist_data[1], bins=(hist_range, hist_range))
H = scipy.ndimage.gaussian_filter(H, sigma=5)
X, Y = np.meshgrid(xedges[:-1] + interval/2, yedges[:-1] + interval/2)
contours = axes.contour(X, Y, H.T, colors='white', linewidths=1, extent=[-0.5,0.5,-0.5,0.5], alpha=0.8, levels=3)

plt.savefig(f"Q2Q3_1000K.png",transparent=True,bbox_inches="tight",dpi=300)
```

```python
matplotlib.rcParams['axes.linewidth'] = 1.5

Q2_arrays_selected = np.array(sorted(Q2_arrays_1000[:, random.randint(low=0, high=len(Q2_arrays_1000[0]), size=20)].T, key=lambda x: x[0]))
print(Q2_arrays_selected.shape)

plt.figure(figsize=(4, 3))
ax = sns.heatmap(Q2_arrays_selected, cmap='bwr', center=0, cbar_kws={"label": "Q2 (Å)"}, linecolor='black', vmin=-0.5, vmax=0.5, xticklabels=5, yticklabels=False)
plt.xlabel("Time (fs)")

plt.gca().spines['top'].set_visible(True)
plt.gca().spines['right'].set_visible(True)
plt.gca().spines['left'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)

ax.tick_params(axis='both', labelsize=8, width=1.5, length=5)
ax.set_xticks(np.arange(0, 120, 20))
ax.set_xticklabels(np.arange(0, 120, 20))

cbar = ax.collections[0].colorbar
for spine in cbar.ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_edgecolor('black')
cbar.ax.tick_params(labelsize=8, width=1.5, length=5)

plt.savefig(f"evolution_1000K.png",transparent=True,bbox_inches="tight",dpi=300)
plt.show()
```
