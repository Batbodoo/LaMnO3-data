```python
import matplotlib.pyplot as plt
import pandas as pd

steps = np.linspace(200,1200,len(a_params))

plt.plot(steps, a_params, color = 'black', alpha=0.2, zorder=1)
plt.plot(steps, b_params, color = 'black', alpha=0.2, zorder=1)
plt.plot(steps, c_params, color = 'black', alpha=0.2, zorder=1)

roll_a = pd.Series(a_params).rolling(100, min_periods=0, win_type="hamming").mean()
roll_b = pd.Series(b_params).rolling(100, min_periods=0, win_type="hamming").mean()
roll_c = pd.Series(c_params).rolling(100, min_periods=0, win_type="hamming").mean()

plt.plot(steps, roll_a,zorder=2, color = 'black', label = 'MLMD')
plt.plot(steps, roll_b,zorder=2, color = 'black')
plt.plot(steps, roll_c,zorder=2, color = 'black')

plt.xlabel('Temperature (K)')
plt.ylabel(r'Reduced cell parameter ($\AA$)')

plt.yticks([5.5, 5.6, 5.7, 5.8])
plt.xticks([400, 600, 800, 1000])

plt.xlim(300, 1000)
plt.ylim(5.4, 5.85)
plt.legend(frameon = False)
plt.savefig("lattice_param_evolution_MLMD_v3.png")
plt.show()
```
